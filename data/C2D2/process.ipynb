```python
import pandas as pd

df = pd.read_csv("C2D2_combined.csv")
df['Distortion Type'].value_counts()
```

```python
mapping = {
    "Non-distorted": "Non-distorted",
    "Mind Reading": "Jumping to Conclusions",
    "Fortune-telling": "Jumping to Conclusions",
    "Overgeneralization": "Overgeneralization",
    "Emotional Reasoning": "Emotional Reasoning",
    "Labeling and Mislabeling": "Labeling and Mislabeling",
    "Personalization": "Personalization",
    "All-or-Nothing Thinking": "All-or-Nothing Thinking"
}

df["Distortion Type"] = df["Distortion Type"].map(mapping)

df["Distortion Type"].value_counts()
```

```python
target_labels = [
    "Emotional Reasoning",
    "Overgeneralization",
    "Should Statements",
    "Personalization",
    "Mental Filter",
    "Disqualifying the Positive",
    "Jumping to Conclusions",
    "Labeling and Mislabeling",
    "Magnification and Minimization",
    "All-or-Nothing Thinking"
]

# Giả sử df là DataFrame gốc
# Tạo cột mới với giá trị mặc định là 0
for label in target_labels:
    df[label] = (df["Distortion Type"] == label).astype(int)

df["Label"] = df["Label"].fillna(
    df["Distortion Type"].apply(lambda x: 0 if x == "Non-distorted" else 1)
).astype(int)


# Chọn và sắp xếp lại các cột theo yêu cầu
df_new = df[["Original Text", "Label"] + target_labels]
```

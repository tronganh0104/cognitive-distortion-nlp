```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("dialogue.csv")
df
```

```python
print("Counts per bias_type (full):")
print(df['bias_type'].value_counts())
print("\nCounts per label:")
print(df['label'].value_counts())
```

```python
df['bias_type'] = df['bias_type'].fillna("None").astype(str)

# --- Train / temp ---
train_df, temp_df = train_test_split(
    df,
    test_size=0.2,
    stratify=df['bias_type'],
    random_state=42
)

# --- Temp -> val + test ---
val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    stratify=temp_df['bias_type'],
    random_state=42
)
```

```python
train_df.to_csv("train.csv", index=False)
val_df.to_csv("val.csv", index=False)
test_df.to_csv("test.csv", index=False)
```

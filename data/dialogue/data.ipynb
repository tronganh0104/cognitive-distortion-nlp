```python
import pandas as pd

dftrain = pd.read_csv('train.csv')
dftest = pd.read_csv('test.csv')
```

```python
df_all = pd.concat([dftrain, dftest], ignore_index=True)
df_all
```

```python
df_all['bias_type'].value_counts()
```

```python
import matplotlib.pyplot as plt

df = df_all
```

```python

counts = df['bias_type'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(counts.index, counts.values)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Số lượng')
plt.title('Phân bố từng nhãn trong bộ dữ liệu')
plt.tight_layout()
plt.show()
```

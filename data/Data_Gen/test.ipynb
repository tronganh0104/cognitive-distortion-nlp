```python
import json
import pandas as pd

file_path = 'dialogue.json'

with open(file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

sample = data[0]
print("=== Ví dụ một mẫu hội thoại ===")
print("Original Text:", sample["original_text"])
print("Label:", sample["label"])
print("Bias type:", sample["bias_type"])
print("Dialogue:")
for turn in sample["dialogue"]:
    print(f"  {turn['speaker']}: {turn['text']}")
print("Explanation:", sample["explanation"])
```

```python
from collections import Counter

print("Tổng số mẫu trong file:", len(data))

bias_counts = Counter(item['bias_type'] for item in data)
print("Số lượng mẫu theo từng loại sai lệch:")
for bias, count in bias_counts.items():
    print(f"- {bias}: {count}")
```

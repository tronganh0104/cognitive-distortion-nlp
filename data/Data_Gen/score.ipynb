```python
import json
import pandas as pd

# ƒê∆∞·ªùng d·∫´n file JSON ƒë·∫ßu v√†o
input_file = "score_data.json"
# ƒê∆∞·ªùng d·∫´n file CSV ƒë·∫ßu ra
output_file = "data_scored.csv"

# ƒê·ªçc file JSON
with open(input_file, "r", encoding="utf-8") as f:
    data = json.load(f)

# N·∫øu file ch·ªâ ch·ª©a 1 dict, ta cho v√†o list ƒë·ªÉ t·∫°o DataFrame
if isinstance(data, dict):
    data = [data]

# T·∫°o DataFrame
df = pd.DataFrame(data)

# Xu·∫•t ra file CSV
df.to_csv(output_file, index=False, encoding="utf-8-sig")

print(f"‚úÖ ƒê√£ chuy·ªÉn {len(df)} d√≤ng d·ªØ li·ªáu sang file '{output_file}'")
```

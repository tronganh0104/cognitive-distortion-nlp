```python
import pandas as pd

df = pd.read_csv("PsyQA_Label_CogDis.csv")
df
```

```python
df["Distortion Type"].value_counts()
```

```python
target_labels = [
    "Emotional Reasoning",
    "Overgeneralization",
    "Should Statements",
    "Personalization",
    "Mental Filter",
    "Disqualifying the Positive",
    "Jumping to Conclusions",
    "Labeling and Mislabeling",
    "Magnification and Minimization",
    "All-or-Nothing Thinking"
]

# Tạo one-hot encoding cho cột Distortion Type
df_onehot = pd.get_dummies(df['Distortion Type'])

for label in target_labels:
    if label not in df_onehot.columns:
        df_onehot[label] = 0

# Chỉ giữ lại và sắp xếp đúng thứ tự các cột
df_onehot = df_onehot[target_labels]

# Gộp lại với cột Original Text và Label
final_df_label1 = pd.concat([df[['Original Text', 'Label']], df_onehot], axis=1)
```

```python
label_cols = [
    "Emotional Reasoning",
    "Overgeneralization",
    "Should Statements",
    "Personalization",
    "Mental Filter",
    "Disqualifying the Positive",
    "Jumping to Conclusions",
    "Labeling and Mislabeling",
    "Magnification and Minimization",
    "All-or-Nothing Thinking"
]

# Chuyển True/False sang 1/0 cho các cột đó
final_df_label1[label_cols] = final_df_label1[label_cols].astype(int)
```

```python
final_df_label1
```

```python
df2 = pd.read_csv("label0.csv")
distortion_cols = [
    "Emotional Reasoning",
    "Overgeneralization",
    "Should Statements",
    "Personalization",
    "Mental Filter",
    "Disqualifying the Positive",
    "Jumping to Conclusions",
    "Labeling and Mislabeling",
    "Magnification and Minimization",
    "All-or-Nothing Thinking"
]

# Tạo các cột mới, tất cả đều = 0
for col in distortion_cols:
    df2[col] = 0

# Sắp xếp lại thứ tự các cột
df2 = df2[["Original Text", "Label"] + distortion_cols]
```

```python
df2
```

```python
combined_df = pd.concat([final_df_label1, df2], ignore_index=True)
```

```python
combined_df
```

```python
combined_df.to_csv("PsyQA_Processed.csv", index=False)
```

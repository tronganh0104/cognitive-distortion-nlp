```python
import pandas as pd

Da_df = pd.read_csv("./CrossLinCD/Da_Processed.csv")
PsyQA_df = pd.read_csv("./PsyQA/PsyQA_Processed.csv")
```

```python
Da_df.head(5)
```

```python
PsyQA_df.head(5)
```

```python
combined_df = pd.concat([Da_df, PsyQA_df], ignore_index=True)
combined_df
```

```python
target_labels = [
    'All-or-Nothing Thinking', 'Overgeneralization', 'Mental Filter',
    'Disqualifying the Positive', 'Jumping to Conclusions', 'Personalization',
    'Emotional Reasoning', 'Should Statements', 'Labeling and Mislabeling',
    'Magnification and Minimization'
]

for label in target_labels:
    if label in combined_df.columns:
        count = combined_df[label].sum()
        print(f"{label}: {count} mẫu")
    else:
        print(f"{label}: 0 mẫu (chưa có dữ liệu)")
```

```python
import pandas as pd

df = pd.read_csv('Da_Trans.csv')
df

original_labels = [
    'Emotional Reasoning', 'Overgeneralization', 'Labeling', 'Mind Reading',
    'Fortune Telling', 'All-or-Nothing Thinking', 'Should Statements',
    'Magnification', 'Personalization', 'Mental Filter', 'Disqualifying the Positive',
    'Blaming Others'
]

target_labels = [
    'All or Nothing Thinking', 'Overgeneralization', 'Mental Filter',
    'Disqualifying the Positive', 'Jumping to Conclusions', 'Personalization',
    'Emotional Reasoning', 'Should Statements', 'Labeling and Mislabeling',
    'Magnification and Minimization'
]

missing_labels = [label for label in original_labels if label not in df.columns]
if missing_labels:
    print(f"Warning: These labels are missing in the dataset: {missing_labels}")
```

```python
# Bước 1: Hợp nhất Mind_Reading và Fortune_Telling thành Jumping_to_Conclusions
df['Jumping to Conclusions'] = df['Mind Reading'] | df['Fortune Telling']

# Bước 2: Ánh xạ các nhãn khớp trực tiếp
# Đổi tên Labeling thành Labeling_and_Mislabeling
df['Labeling and Mislabeling'] = df['Labeling']
# Các nhãn còn lại giữ nguyên tên
direct_mappings = {
    'Emotional Reasoning': 'Emotional Reasoning',
    'Overgeneralization': 'Overgeneralization',
    'All-or-Nothing Thinking': 'All or Nothing Thinking',
    'Should Statements': 'Should Statements',
    'Mental Filter': 'Mental Filter',
    'Disqualifying the Positive': 'Disqualifying the Positive',
    'Magnification and Minimization': 'Magnification and Minimization'
}
for original, target in direct_mappings.items():
    df[target] = df[original]

# Bước 3: Loại bỏ các nhãn không cần thiết
labels_to_drop = ['Mind Reading', 'Fortune Telling', 'Self-Blame', 'Labeling', 'Magnification', 'Blaming Others', 'All-or-Nothing Thinking']
df = df.drop(columns=labels_to_drop, errors='ignore')

# Bước 4: Kiểm tra phân bố nhãn
print("Phân bố nhãn sau chuẩn hóa:")
for label in target_labels:
    if label in df.columns:
        count = df[label].sum()
        print(f"{label}: {count} mẫu")
    else:
        print(f"{label}: 0 mẫu (chưa có dữ liệu)")
    
```

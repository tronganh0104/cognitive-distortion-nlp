```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
binary_df = pd.read_csv('feature_model_comparison_results_binary.csv')

plt.figure(figsize=(10,6))
sns.barplot(data=binary_df, x='Feature Set', y='F1-score', hue='Model')
plt.title('Binary Classification: F1-score theo Feature Set và Model')
plt.ylabel('F1-score')
plt.ylim(0,1)
plt.show()
```

```python
multi_df = pd.read_csv('feature_vs_each_cognitive_bias.csv')

# Pivot để heatmap
heatmap_df = multi_df.pivot_table(index='Target', columns='Feature', values='F1-macro', aggfunc='mean')

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_df, annot=True, fmt=".2f", cmap='YlGnBu')
plt.title('F1-macro theo Feature Set và Cognitive Bias Label')
plt.ylabel('Cognitive Bias Label')
plt.xlabel('Feature Set')
plt.show()
```

```python
for target in multi_df['Target'].unique():
    plt.figure(figsize=(10,5))
    df_plot = multi_df[multi_df['Target']==target]
    sns.barplot(data=df_plot, x='Feature', y='F1-macro', hue='Model')
    plt.title(f'F1-macro cho nhãn "{target}"')
    plt.ylim(0,1)
    plt.ylabel('F1-macro')
    plt.show()
```
